```python
# Joins
```

```python
grade_lookup = {'Grade Level': pd.Series(['ELEMENTARY', 'MIDDLE/HIGH', 'MISC']),  'Level': pd.Series([1, 2, 3])}
grade_lookup = pd.DataFrame(grade_lookup)
grade_lookup[['Grade Level']][0:5]
```

```python
d_sub = d[0:5].join(grade_lookup.set_index(['Grade Level']),on=['Grade Level'], how='inner')
d_sub[['Grade Level', 'Level']]
```

```python
d_sub = d[0:5].join(grade_lookup.set_index(['Grade Level']),on=['Grade Level'], how='left')
d_sub[['Grade Level','Level']]
```

```python
d_sub = d[0:5].join(grade_lookup.set_index(['Grade Level']),on=['Grade Level'], how='outer')
d_sub[['Grade Level','Level']]
```
