```python
# Stemming and lemmatization
## Stemming is a process of reducing a word into its root form.
## root form is not a word by itself,but words can be formed by adding the right suffix to it
## if you take fish, fishes and fishing..they all can be stemmed to fishing
## also, study, studyies and studying can be stemmed to study
## porterstemmer, lancaster, snowball, and so on.
```

```python
## porter is the most commonly used stemmer.
## It is also one of the gentlest stemmers and is computationally intensive with regard to algorithms
```

```python
## The snowball algorithm is regarded as an improvement over porter
```

```python
## Lancaster is a more aggressive stemming algorithm.
## Porter and Snowball stemming is understandable by readers, but lancaster is not. as it makes
## words more obscure. Lancaster is considered to be the fastet algorithm among the three
## and it will work very well with a large set of words,
## but if you are looking for something more distinctive, then Lancaster is not for you.
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
for w in word_data[:20]:
    print ("Actual: %s Stem: %s" %(w,porter_stemmer.stem(w)))
```

```python
from nltk.stem.lancaster import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
```

```python
for w in word_data[:20]:
    print("Actual: %s Stem: %s" %(w,lancaster_stemmer.stem(w)))
```

```python
from nltk.stem.snowball import SnowballStemmer
snowball_stemmer = SnowballStemmer("english")
for w in word_data[:20]:
    print ("Actual: %s Stem: %s" %(w,snowball_stemmer.stem(w)))
```

```python
## Lemmatization
### Lemmatization is similar to stemming but unlike stemming, it brings
### in a context of the word
### e.g: A lemmatization-based algorithm will match a train to the word locomotive,
### but a stemming algorithm won't be able to do this. It makes use of
### a dictionary to link up words

### The WordNet is a lexical database for English by princeton, 
### and we'll use their lemmatization techniques
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
for w in word_data[:20]:
    print("Actual: %s lemm: %s" %(w, lemmatizer.lemmatize(w)))
```
